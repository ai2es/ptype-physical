```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import yaml
import glob
import tqdm
from plotting import ptype_hist
```

```python
config = 'config.yml'
with open(config) as f:
    conf = yaml.load(f, Loader=yaml.FullLoader)

input_cols = conf['tempvars'] + conf['tempdewvars'] + conf['ugrdvars'] + conf['vgrdvars']
output_cols = conf['outputvars']

df = pd.read_parquet(conf['data_path'])
```

```python
fig, ax = plt.subplots(dpi=200, figsize=(6,4))
ax.hist(df['ra_percent'][df['ra_percent']> 0]*100, log=True)
ax.set_title('mPING ra > 0', fontsize=16)
ax.set_xlabel('Percent of Observations', fontsize=12)
ax.set_ylabel('Count', fontsize=12)
```

```python
bins = np.arange(-30, 30, 1)
fig, ax = plt.subplots(dpi=200, figsize=(6,4))
ax.hist(df['TEMP_C_0_m'][df['sn_percent']> 0], bins=bins)
ax.set_title('ASOS sn Observations vs. Temp (0 m)', fontsize=16)
ax.set_xlabel('Temp 0 m ($\degree$C)', fontsize=12)
ax.set_ylabel('Count', fontsize=12)
```
