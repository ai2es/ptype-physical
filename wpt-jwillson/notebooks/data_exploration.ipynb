# Imports and Function Definitions

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import yaml
import glob
import tqdm
from plotting import ptype_hist
```

# ASOS Data Exploration

### Get features and outputs from config.yml

```python
config = 'config.yml'
with open(config) as f:
    conf = yaml.load(f, Loader=yaml.FullLoader)

input_cols = conf['tempvars'] + conf['tempdewvars'] + conf['ugrdvars'] + conf['vgrdvars']
output_cols = conf['outputvars']
```

```python
# sort files by their date
arr = sorted(glob.glob(conf['asos_path']+'*'))
```

### Load all ASOS data

```python
df = pd.read_parquet(conf['data_path'])
```

```python
df = pd.concat([pd.read_parquet(x) for x in tqdm.tqdm(arr)])
df.shape
```

### Make histograms

```python
plt.hist(df['ra_percent'])
plt.title('ASOS ra')
```

```python
plt.hist(df['ra_percent'][df['ra_percent']> 0])
plt.title('ASOS ra > 0')
```

```python
plt.hist(df['sn_percent'])
plt.title('ASOS sn')
```

```python
plt.hist(df['sn_percent'][df['sn_percent']> 0])
plt.title('ASOS sn > 0')
```

```python
plt.hist(df['pl_percent'])
plt.title('ASOS pl')
```

```python
plt.hist(df['pl_percent'][df['pl_percent']> 0])
plt.title('ASOS pl > 0')
```

```python
plt.hist(df['fzra_percent'])
plt.title('ASOS fzra')
```

```python
plt.hist(df['fzra_percent'][df['fzra_percent']> 0])
plt.title('ASOS fzra > 0')
```

### Get conditional dists

```python
ra = df[input_cols][df['ra_percent'] > 0]
sn = df[input_cols][df['sn_percent'] > 0]
pl = df[input_cols][df['pl_percent'] > 0]
frza = df[input_cols][df['fzra_percent'] > 0]
```

```python
bins=np.arange(-40,40,1)
ptype_hist(df, 'TEMP_C_0_m', 'ASOS', bins=bins)
```

```python
ptype_hist('TEMP_C_1000_m', 'ASOS', bins=bins)
```

```python
ptype_hist('T_DEWPOINT_C_0_m', 'ASOS', bins=bins)
```

```python
bins = np.arange(-40,40,1)
ptype_hist('UGRD_m/s_1000_m', 'ASOS', bins=bins)
```

```python
bins=np.arange(-15,15,1)
ptype_hist('VGRD_m/s_0_m', 'ASOS', bins=bins)
```

```python
bins = np.arange(-40,40,1)
ptype_hist('VGRD_m/s_1000_m', 'ASOS', bins=bins)
```
